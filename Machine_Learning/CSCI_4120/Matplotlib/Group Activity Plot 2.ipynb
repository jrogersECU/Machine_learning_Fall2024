```python
import pandas as pd
# TODO Import plot lib
from matplotlib import pyplot as plt

%matplotlib inline
```

```python
df =  pd.read_csv('iris_data.csv', sep=',')
print(df[:5])
```

```python

# TODO visualize petal length (x-axis) vs petal width (y-axis) using a scatter chart
# the color and size of each data record are decided by 100* petal length * petal width
# you also need a color bar, alpha=0.3
# you may need to check this website: https://pandas.pydata.org/pandas-docs/stable/user_guide/indexing.html 
# to learn how to get the data from a dataframe (df).
plt.scatter(df['petal_length'], df['petal_width'], c=100*df['petal_length']*df['petal_width'], s=100*df['petal_length']*df['petal_width'], alpha=0.3, cmap='viridis')
plt.xlabel('petal_length')
plt.ylabel('petal_width')
plt.title('petal_length vs petal_width')
plt.colorbar()
```
